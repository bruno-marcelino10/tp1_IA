```python
import sys

class Entrada:
    values: list
    nome_binario: str
    nome_trabalho: str
    nome_algoritmo: str
    printar: bool

    def __init__(self) -> None:
        self.values = sys.argv
        self.nome_binario = self.values.pop(0) # Recebendo o nome do binário da lista da entrada
        self.nome_trabalho = self.values.pop(0) # Recebendo o nome do trabalho
        self.nome_algoritmo = self.values.pop(0) # Recebendo o nome do algoritmo
        self.printar = False

        try:
            int(self.values[-1]) # Verificando se o último valor da lista é a palavra print
        except:
            self.values.pop(-1)
            self.printar = True

        map(int, self.values)

    def __str__(self) -> str:
        buffer = "Binário: " + self.nome_binario + "\nNome do trabalho: " + self.nome_trabalho +\
            "\nNome do algoritmo: " + self.nome_algoritmo + "\nValores: ["

        i = 0
        for valor in self.values:
            buffer += f"{valor}"
            i += 1

            if i < len(self.values):
                buffer += ", "

        buffer += f"]\nDevo printar passo a passo na tela? {self.printar}."

        return buffer
```

## Busca em Largura

```python
def action(node):
    # node = [estado, pai, ação, custo, profundidade]
    lista = []
    d = node[4] + 1
    custo = node[3] + 1
    for i in range(len(node[0])):
        for j in range(len(node[0])):
            if node[0][i] > node[0][j]:
                
                estado = node[0].copy()
                tmp = estado[i]
                estado[i] = estado[j]
                estado[j] = tmp

                lista.append([estado, node[0], [i+1, j+1], custo, d])

    return lista
```

```python
def bfs(init_node: list):

    init_node = [init_node, init_node, 0, 0, 0]
    frontier = None
    explored = []
    solution = init_node[0].copy()
    solution.sort()
    
    while True:

        # Primeiro caso
        if frontier == None:
            frontier = action(init_node)
            explored = [init_node[0]]
            print("Iniciando...")
            print("Input: [estado, pai, ação (trocas), custo, profundidade]")
            print(init_node)
        else:
            # Caso onde não há solução
            if frontier == []:
                return "Sem solução"
            
            # Caso onde há solução
            else:

                # Seleciona um elemento
                explored_node = frontier[0]

                # Teste da solução
                if explored_node[0] == solution:
                    print(explored_node)
                    return "Fim"
                
                # Caso onde nó já foi explorado
                if explored_node[0] in explored:
                    frontier.remove(explored_node)

                # Caso onde nó ainda não foi explorado
                else:
                    print(explored_node)
                    frontier = frontier + action(explored_node)
                    explored.append(explored_node[0])
                    frontier.remove(explored_node)
```

```python
def call_bfs(argv_input):
    entrada = entrada(argv_input)
    bfs(entrada)
```

## Busca Com Aprofundamento Iterativo

```python
def action(node):
    # node = [estado, pai, ação, custo, profundidade]
    lista = []
    d = node[4] + 1
    custo = node[3] + 1
    for i in range(len(node[0])):
        for j in range(len(node[0])):
            if node[0][i] > node[0][j]:
                
                estado = node[0].copy()
                tmp = estado[i]
                estado[i] = estado[j]
                estado[j] = tmp

                lista.append([estado, node[0], [i+1, j+1], custo, d])

    return lista
```

```python
def depth_node(frontier: list):
    # node = [estado, pai, ação, custo, profundidade]
    depths = [node[4] for node in frontier]
    for i in range(len(depths)):
        if depths[i] == max(depths):
            chosen_node = frontier[i]
            break
    
    return chosen_node
```

```python
def dfs(init_node: list, max_depth: int):

    init_node = [init_node, init_node, 0, 0, 0]
    frontier = None
    explored = []
    solution = init_node[0].copy()
    solution.sort()
    
    while True:

        # Primeiro caso
        if frontier == None:
            frontier = action(init_node)
            explored = [init_node[0]]
            print("\nIniciando...")
            print("Profundidade:", max_depth)
            print(init_node)
        else:
            # Caso onde não há solução
            if frontier == []:
                return ["Sem solução"]
            
            # Caso onde há solução
            else:

                # Seleciona um elemento
                explored_node = depth_node(frontier)

                # Teste da solução
                if explored_node[0] == solution:
                    print(explored_node)
                    return ["Fim", explored_node]
                
                # Caso onde nó já foi explorado ou está na última profundidade
                if explored_node[4] == max_depth or explored_node[0] in explored:
                    frontier.remove(explored_node)

                # Caso onde nó ainda não foi explorado
                else:
                    print(explored_node)
                    frontier = frontier + action(explored_node)
                    explored.append(explored_node[0])
                    frontier.remove(explored_node)
```

```python
def ids(init_node: list, depth: int):
    for i in range(1, depth+1):
        iter_dfs = dfs(init_node, i)
        
        if iter_dfs[0] == "Fim":
            print("\nSolução:", iter_dfs[1])
            return None
    
    return "sem solução"
```

## Busca de Custo Uniforme (Dijkstra)

```python
def action(node):
    # node = [estado, pai, ação, custo, profundidade]
    lista = []
    d = node[4] + 1

    for i in range(len(node[0])):
        for j in range(len(node[0])):
            if node[0][i] > node[0][j]:
                
                estado = node[0].copy()
                tmp = estado[i]
                estado[i] = estado[j]
                estado[j] = tmp
                
                if j-i == 1:
                    custo = 2
                else:
                    custo = 4

                lista.append([estado, node[0], [i+1, j+1], node[3] + custo, d])

    return lista
```

```python
def low_cost_node(frontier: list):
    # node = [estado, pai, ação, custo, profundidade]
    costs = [node[3] for node in frontier]
    for i in range(len(costs)):
        if costs[i] == min(costs):
            chosen_node = frontier[i]
            break
    
    return chosen_node
```

```python
def ucs(init_node: list):

    init_node = [init_node, init_node, 0, 0, 0]
    frontier = None
    explored = []
    solution = init_node[0].copy()
    solution.sort()
    
    while True:

        # Primeiro caso
        if frontier == None:
            frontier = action(init_node)
            explored = [init_node[0]]
            print("\nIniciando...")
            print("Input: [estado, pai, ação (trocas), custo, profundidade]")
            print(init_node)
        else:
            # Caso onde não há solução
            if frontier == []:
                return ["Sem solução"]
            
            # Caso onde há solução
            else:

                # Seleciona um elemento
                explored_node = low_cost_node(frontier)

                # Teste da solução
                if explored_node[0] == solution:
                    print(explored_node)
                    return ["Fim", explored_node]
                
                # Caso onde nó já foi explorado ou está na última profundidade
                if explored_node[0] in explored:
                    frontier.remove(explored_node)

                # Caso onde nó ainda não foi explorado
                else:
                    print(explored_node)
                    frontier = frontier + action(explored_node)
                    explored.append(explored_node[0])
                    frontier.remove(explored_node)
```

## Busca A*

```python
# Heurística: Qtde numeros fora do lugar
def action(node):
    # node = [estado, pai, ação, heuristica, custo, f(n), profundidade]
    lista = []
    d = node[6] + 1

    for i in range(len(node[0])):
        for j in range(len(node[0])):
            if node[0][i] > node[0][j]:
                
                # Cria novo estado
                estado = node[0].copy()
                tmp = estado[i]
                estado[i] = estado[j]
                estado[j] = tmp
                
                # Calcula custo total
                if j-i == 1:
                    custo = 2
                else:
                    custo = 4
                
                # Calcula h(n) -> Qtde numeros fora do lugar
                h = 0
                for pos in range(len(estado)): # [2,3,1,5,4]
                    solution = estado.copy()
                    solution.sort() # [1,2,3,4,5]
                    
                    best_pos = estado[pos]-1
                    
                    if best_pos != pos:
                        h += 1
                
                lista.append([estado, node[0], [i+1, j+1], h, node[4] + custo, node[4] + custo + h,d])

    return lista
```

```python
def low_f_node(frontier: list):
    # node = [estado, pai, ação, heuristica, custo, f(n), profundidade]
    fs = [node[5] for node in frontier]
    for i in range(len(fs)):
        if fs[i] == min(fs):
            chosen_node = frontier[i]
            break
    
    return chosen_node
```

```python
def a_star(init_node: list):

    init_node = [init_node, init_node, 0, 0, 0, 0, 0]
    frontier = None
    explored = []
    solution = init_node[0].copy()
    solution.sort()
    
    while True:

        # Primeiro caso
        if frontier == None:
            frontier = action(init_node)
            explored = [init_node[0]]
            print("\nIniciando...")
            print("Input: [estado, pai, ação, heuristica, custo, f(n), profundidade]")
        else:
            # Caso onde não há solução
            if frontier == []:
                return ["Sem solução"]
            
            # Caso onde há solução
            else:

                # Seleciona um elemento
                explored_node = low_f_node(frontier)

                # Teste da solução
                if explored_node[0] == solution:
                    print(explored_node)
                    return ["Fim", explored_node]
                
                # Caso onde nó já foi explorado ou está na última profundidade
                if explored_node[0] in explored:
                    frontier.remove(explored_node)

                # Caso onde nó ainda não foi explorado
                else:
                    print(explored_node)
                    frontier = frontier + action(explored_node)
                    explored.append(explored_node[0])
                    frontier.remove(explored_node)
```
